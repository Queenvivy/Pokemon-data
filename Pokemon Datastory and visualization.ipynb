# Data Analysis and Visualizations


```python
#to check the stat that correlates with the other
data.corr()
```

```python
fig,axes = plt.subplots(figsize=(14,8))
sns.heatmap(data.corr(),annot=True).set_title('General Stats')
plt.show()
```

##### here in general stats we can see that:
>Total has a good correlation with the attack and defence attributes. For example,

>Total with attack: 0.73

>Total with sp.attack: 0.72

>Total with defense: 0.64

>Total with sp.defense: 0.72

>The statement suggests that the "Total" attribute of PokÃ©mon has a strong positive correlation with both the attack and defense attributes. The correlation coefficients provided indicate the strength and direction of the relationship between the "Total" attribute and each specific attribute.

>In this case, a correlation coefficient of 0.73 between the "Total" attribute and attack attribute suggests a relatively strong positive relationship. Similarly, a correlation coefficient of 0.72 between the "Total" attribute and sp.attack (special attack) as well as sp.defense (special defense) attributes also indicates a strong positive relationship.

>However, the correlation coefficient of 0.64 between the "Total" attribute and defense attribute suggests a slightly weaker positive relationship compared to the other attributes.

>Overall, these correlations suggest that as the "Total" attribute of a PokÃ©mon increases, there tends to be a corresponding increase in the attack, special attack, and special defense attributes. The defense attribute, while still positively correlated, may have a slightly weaker relationship with the "Total" attribute.


```python
#How speed of Pokemon relates to various base factors?

fig,axes = plt.subplots(2,2,figsize=(16,10),sharey=True)
sns.scatterplot(data['Attack'],data['Speed'],ax=axes[0,0])
axes[0,0].set_title("Speed V/S Attack")
sns.scatterplot(data['Defense'],data['Speed'],ax=axes[0,1])
axes[0,1].set_title("Speed V/S Defence")
sns.scatterplot(data['HP'],data['Speed'],ax=axes[1,0])
axes[1,0].set_title("Speed V/S HP")
fig.suptitle("Speed Factor", size=20)
plt.show()
```

##### Some insights that we can get from the above graphs are:
  
  ##### ATTACK
>A pokemon's attack capacity slightly depends on its speed
>some of the pokemons like mewtwo mega mewtwo X have moderate speed but a good attack capacity,
>some are very fast yet lacking in their attack capacity.

   ##### DEFENSE
>For a pokemon's defense speed is not that much required 
>from the above Speed vs Defence plot that, pokemon's which are slower possess more defense capacity.
>e.g Eternatus Eternamax has high defense of 250 but low speed of 130

   ##### HITPOINT
>the hitpoint of a pokemon highly affects its speed
>from the speed vs HP plot, pokemons with high HP possess less speed
>e.g Blissey with high HP of 255 has a speed of 55.


```python
#What are the most wide spread types of pokemon



fig,axes = plt.subplots(2,2,figsize=(16,9))
ax = sns.countplot(x='Type1',data=data,order=data['Type1'].value_counts().iloc[:5].index, ax=axes[0,0])
ax.set_title('Type1 pokemon with the widest spread')
for p in ax.patches:
    ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))

ax = sns.countplot(x='Type2',data=data,order=data['Type2'].value_counts().iloc[:5].index, ax=axes[0,1])
ax.set_title('Type2 pokemons with the widest spread')
for p in ax.patches:
    ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))

ax = sns.countplot(x='Type1',data=data,order=data['Type1'].value_counts(ascending=True).index[:5], ax=axes[1,0])
ax.set_title('Type1 pokemons with lowest spread')

for p in ax.patches:
    ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))

ax = sns.countplot(x='Type2',data=data,order=data['Type2'].value_counts(ascending=True).index[:5], ax=axes[1,1])
ax.set_title('Type2 pokemons with lowest spread')
for p in ax.patches:
    ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))
plt.show()
```

##### The most wide spread type of pokemon in:
> Type1 = Water type pokemons

> Type2 = Flying type pokemons (as this is a type2 so most of the pokemon has no type2)

#####  The least spread type of pokemon in:
> Type1 = Flying type pokemons
>Type2 = Bug type pokemons

Note that while there are most number of pokemons with Flying type in thier type2 yet the flying type has the least number in type1.

```python
#Which type is the strongest overall?

top5_pokemon_Total = data.sort_values(by="Total", ascending=False).reset_index()[:5]
plt.figure(figsize=(15,8))
ax = sns.barplot(x=top5_pokemon_Total["Names"], y=top5_pokemon_Total["Total"], orient='v')
ax.set_title("Which is the best pokÃ©mon?", size=20)
ax.set(xlabel="Total", ylabel="Names")
for p in ax.patches:
    ax.annotate('{:.1f}'.format( p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))
    
    
```

From above plot,comparing all the base strengths,we have Eternatus Eternamax at highest strength,followed by mewtwo mega mewtwo X.
To determine the strongest Pokemon, we summed up the base statistics for each species. 
While our metric approached Pokemon strength objectively as possible, in the end, it may not be meaningful to all players.

Therefore, the player's selection of moves, attack and defense points during the battle is ultimately a bigger factor 
in determining success than the strength of the Pokemon chosen. And hence the developers have made Pokemon a game that is more 
than simply optimizing statistics and strength, but is a tactical and personal experience.
